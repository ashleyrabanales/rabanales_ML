```python
#novel activation 
import numpy as np
import matplotlib.pyplot as plt

def my_activation_function(x):
    # Define the activation function here
    return 1 / (2 + np.exp(-x)) - 0.6 * np.cos(x)

# Generate x values
x = np.linspace(-8, 10, 120)

# Evaluate the activation function for each x value
y = my_activation_function(x)

# Create a plot of the activation function
plt.plot(x, y)
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('My Activation Function')
plt.show()
```

- The wave-like activation function, which utilizes cosine, serves the purpose of capturing both non-linear and periodic patterns in data. 

- While the sigmoid function is adept at detecting non-linear patterns and the sine function is effective at identifying periodic patterns, the wave-like activation function is particularly advantageous for specific problem types, including signal processing and time-series prediction.
